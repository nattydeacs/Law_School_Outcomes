```python
import pandas as pd
import numpy as np
```

```python
#read in employment files
years = list(range(2013, 2023))
employmentTables = []
for year in years:
    fileName = ["ABA_downloads/Employment Summary_", str(year), ".xlsx"]
    fileName = "".join(fileName)
    file = pd.read_excel(fileName)
    employmentTables.append(file)
```

```python
year = 2012
for i in range(0,10):
    employmentTables[i]["Year"] = year
    year += 1
```

```python
allYears = pd.concat(employmentTables[0:10])
```

```python
#reshaping
FirstStates = allYears[['SchoolName', 'Year',
             'FirstLargestEmployment','FirstLargestEmploymentNumber']].copy()
FirstStates["state_rank"] = 1
FirstStates = FirstStates.rename(columns = {'FirstLargestEmployment': 'State', 
                                                'FirstLargestEmploymentNumber': 'number_grads'})

secondStates = allYears[['SchoolName', 'Year',
             'SecondLargestEmployment','SecondLargestEmploymentNumber']].copy()
secondStates["state_rank"] = 2
secondStates = secondStates.rename(columns = {'SecondLargestEmployment': 'State', 
                                                  'SecondLargestEmploymentNumber': 'number_grads'})

ThirdStates = allYears[['SchoolName', 'Year',
             'ThirdLargestEmployment','ThirdLargestEmploymentNumber']].copy()
ThirdStates["state_rank"] = 3
ThirdStates = ThirdStates.rename(columns = {'ThirdLargestEmployment': 'State', 
                                                  'ThirdLargestEmploymentNumber': 'number_grads'})

states = pd.concat([FirstStates, secondStates, ThirdStates])
```
